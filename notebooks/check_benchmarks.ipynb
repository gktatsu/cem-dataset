```python
import os
import numpy as np
from glob import glob
from skimage import io
from matplotlib import pyplot as plt
```

```python
#root_dir is the directory to which the benchmarks were saved
#when the evaluation/setup_benchmarks/setup_data.py script was run, we'll assume it's the 
#default save directory for that script
root_dir = '../evaluation/benchmark_data/'
```

```python
#rather than show examples for all of them simultaneously
#just chose one from the list and comment out the others

#benchmark = 'lucchi_pp'
#benchmark = 'kasthuri_pp'
#benchmark = 'perez/mito'
#benchmark = 'perez/lyso'
#benchmark = 'perez/nuclei'
#benchmark = 'perez/nucleoli'
#benchmark = 'guay/2d'
benchmark = 'urocell/2d'
#benchmark = 'cremi/2d'
#benchmark = 'all_mito'
```

```python
#some training images are tiffs, others are pngs
#we'll make the glob function agnostic
trn_impaths = np.sort(glob(os.path.join(root_dir, f'{benchmark}/train/images/*')))
trn_mskpaths = np.sort(glob(os.path.join(root_dir, f'{benchmark}/train/masks/*')))
```

```python
len(trn_impaths), len(trn_mskpaths)
```

```python
#plot some of the training set images
f, ax = plt.subplots(2, 4, figsize=(12, 6))

for y in range(2):
    for x in range(4):
        #randomly pick an image and mask
        trn_index = np.random.randint(len(trn_impaths))
        trn_image, trn_mask = io.imread(trn_impaths[trn_index]), io.imread(trn_mskpaths[trn_index])

        ax[y, x].imshow(trn_image, cmap='gray')
        ax[y, x].imshow(trn_mask, alpha=0.3)

plt.tight_layout()
```
